The following notebook will provide us with the opportunity to visualize all the passes done by a player or an entire team during a specific match, using Whoscored processed data. To do so, we will use the modules WS_scrape and WS_pass_arrows, which must be in the same directory as the note.

Before anything, we have to import the following libraries:

# Importing libraries

```python
import WS_pass_arrows as arrows
import WS_scrape as scrape
import json
import pandas as pd
```

# Obtaining data

After that, we are obtaining the match's data that we want. In my case, it will be from the last showdown between Liverpool and Manchester City. We can split the data obtained in:

- match_data (json)
- match_keys (json)
- event_data (df)

```python
url = 'https://www.whoscored.com/Matches/1729398/Live/England-Premier-League-2023-2024-Liverpool-Manchester-City'
key = 'events'
match_data, match_keys, df = scrape.get_data(url,key)
```

# Storing data locally and loading it later

Now, we are going to do a little break down. In order to avoid having to execute the cell above again when we want to restart the project or using the data in another one, we are going to store it locally. To do that, we only have to download the event_data and the match_data with the following code:

### Storing

```python
df.to_csv(r'destination/path/file.csv')
with open(r'destination/path/file.json', 'w') as match_data_json_file:
    json.dump(match_data, match_data_json_file)
```

### Loading

```python
with open(r'destination/path/file.json', 'r') as json_file:
    match_data =  json.load(json_file)
df = pd.read_csv(r'destination/path/file.csv')
```

# Plotting

Once we have data loaded again, we can plot the pass arrows figure:

```python
arrows.plot_arrows(df, match_data, player="Rodri")
```

As we can see, Rodri is the brain of the team, accumulating a large number of passes and showing a wide variety of distributions to both sides of the pitch.
